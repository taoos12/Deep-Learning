```python
def Perceptron(X, y):
    # Inserting bias at the beginning
    # np.insert(X, index, value on index, axis)
    X = np.insert(X, 0, 1, axis= 1)

    # adding weights, X.shape[1] = X row size
    weights= np.ones(X.shape[1])
    
    # Initializing Learning rate 
    learning_rate= 0.1

    for i in range(1000):
        # Getting a random value and storing it into a variable
        j = np.random.randint(0, 100)
        
        # if dot product's >= 0, Placement = Yes 
        y_hat = step(np.dot(X[j], weights))

        # Weight updation code
        weights = weights + learning_rate * (y[j] - y_hat)*X[j]

    return weights[0],weights[1:]
```
