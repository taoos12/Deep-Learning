```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
df.drop(columns= 'city', inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['cgpa'].fillna(df['cgpa'].mean(), inplace= True)
df['iq'].fillna(df['iq'].mean(), inplace= True)
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x= df['iq'],y=  df['cgpa'], hue= df['placement'])
plt.legend()
```

```python
# Splitting into labels and features
# iloc = Selects rows/columns by index position (no by name)
# : means all
# 0:2 from column index 0 upto column index 2(Excluding index no 2)
# x = df.iloc[:, 0:2]

# -1 → in Python means "the last element".
# So this selects the last column of the DataFrame.
# df.iloc[:, -1]

X = df.iloc[:, 0:2]
y = df.iloc[:, -1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X, y)
```

```python
# z = w1x1 + w2x2 + b
# z = 404.75 + 44.58
p.coef_
```

```python
# b = -49
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
# plot_decision_regions(X, y, classification model, legend= Quadrant No)
plot_decision_regions(X.values, y.values, clf= p, legend= 2)
```
